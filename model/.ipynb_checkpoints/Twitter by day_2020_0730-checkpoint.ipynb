```python
import twint
import nest_asyncio
import pandas as pd
```

# 2020

```python
def generate(x)->int:
    ymdlist=list()
    year=str(x)
    for i in range(1,13):
        if i==2:
            month='02'
            daylist=[year+'-'+month+'-'+"%.2d" % i for i in range(1,29)]
            ymdlist.extend(daylist)

        if i in [1,3,5,7,8,10,12]:
            month="%.2d" % i
            daylist=[year+'-'+month+'-'+"%.2d" % i for i in range(1,32)]
            ymdlist.extend(daylist)
         
        if i in [4,6,9,11]:
            month="%.2d" % i
            daylist=[year+'-'+month+'-'+"%.2d" % i for i in range(1,31)]
            ymdlist.extend(daylist)
            
    from_ymdhms_list=list()
    to_ymdhms_list=list()
    for i in ymdlist:
        hour='00'
        minute='00'
        second='00'
        ymdhms=i+' '+hour+':'+minute+':'+second
        from_ymdhms_list.append(ymdhms)
        to_ymdhms_list.append(ymdhms)
    from_ymdhms_list.insert(0,str(x-1)+'-12-31 00:00:00')
    from_ymdhms_list.insert(0,str(x-1)+'-12-30 00:00:00')
    from_ymdhms_list.append(str(x+1)+'-01-01 00:00:00')
    from_ymdhms_list.append(str(x+1)+'-01-02 00:00:00')
    to_ymdhms_list.append(str(x+1)+'-01-01 00:00:00')
    to_ymdhms_list.append(str(x+1)+'-01-02 00:00:00')
    to_ymdhms_list.append(str(x+1)+'-01-03 00:00:00')
    to_ymdhms_list.append(str(x+1)+'-01-04 00:00:00') 
    return from_ymdhms_list,to_ymdhms_list
        
```

```python
from_ymdhms_list=generate(2020)[0]
to_ymdhms_list=generate(2020)[1]
```

```python
import random
import time
aa_2020_5=pd.DataFrame()
for i in range(318, len(from_ymdhms_list)):
    nest_asyncio.apply()
    c = twint.Config()
    c.Search = 'american airlines'
    c.Pandas = True
    c.Hide_output = True
    c.Min_likes = 1
    c.Since = from_ymdhms_list[i]
    c.Until= to_ymdhms_list[i]
    c.Debug = True
    c.Lang = 'en'
    c.Limit=600
    twint.run.Search(c)
    df = twint.storage.panda.Tweets_df
    if len(df)==0:
        print(i,'没有数据')
    else:
        df=df[df['language']=='en']
        df=df[['created_at','date','time','username','name','place','tweet','language','replies_count','retweets_count','likes_count']]
        aa_2020_5=pd.concat([aa_2020_5,df], ignore_index=True)
    q=random.uniform(0,2)
    time.sleep(q)
    print(i)
```
